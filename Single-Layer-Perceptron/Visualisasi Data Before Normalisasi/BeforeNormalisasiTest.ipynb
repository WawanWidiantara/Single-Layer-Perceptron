```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('D:/Matkul/JARINGAN SYARAF TIRUAN/TUGAS KELOMPOK/Data/Data Testing.xlsx')
df
```

```python
plt.figure(figsize=(10, 6))
sns.displot(data=df, x='PassengerId', color='Green')
plt.title('Histogram Data')
plt.xlabel('PassengerId')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Pclass',color='Green')
plt.title('Histogram Data')
plt.xlabel('Pclass')
plt.show()
```

```python
Sex_counts = df['Sex'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(Sex_counts, labels=Sex_counts.index, autopct='%1.1f%%', colors=['Green', 'red'])
plt.title('Histogram Data')
plt.xlabel('Sex')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.displot(data=df, x='Age', kind='kde', color='blue')
plt.title('Histogram Data')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Embarked', color='red')
plt.title('Histogram Data')
plt.xlabel('Embarked')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='SibSp', color='red')
plt.title('Histogram Data')
plt.xlabel('SibSp')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Parch', color='Orange')
plt.title('Histogram Data')
plt.xlabel('Pacrh')
plt.show()
```
