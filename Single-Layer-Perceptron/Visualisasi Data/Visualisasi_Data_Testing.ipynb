```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:/visualisasi data\Artificial-Neural-Network/Single-Layer-Perceptron/data/test_data_cl_v2.csv")
df.head()
```

VISUALISASI DATA

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Pclass',  color='red')
plt.title('Histogram Data')
plt.xlabel('Pclass')
plt.ylabel('Jumlah')
plt.show()
```

```python
sex_counts = df['Sex'].value_counts()

plt.figure(figsize=(10, 6))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', colors=['yellow', 'pink'])
plt.title('Data Sex')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.displot(data=df, x='Age', kind='kde', color='brown')
plt.title('Histogram Data')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='SibSp', color='green')
plt.title('Histogram Data')
plt.xlabel('SibSp')
plt.ylabel('Jumlah')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Parch', color='blue')
plt.title('Histogram Data')
plt.xlabel('Parch')
plt.ylabel('Jumlah')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.displot(data=df, x='Fare', kind='kde', color='red')
plt.title('Histogram Data')
plt.xlabel('Fare')
plt.show()
```
