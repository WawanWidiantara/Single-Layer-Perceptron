```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("D:/visualisasi data/Artificial-Neural-Network/Single-Layer-Perceptron/Data Training.xlsx")
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.displot(data=df, x='PassengerId', kind='kde', color='black', hue='Survived')
plt.title('Histogram Data')
plt.xlabel('PassengerId')
plt.show()
```
