```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("D:/visualisasi data/Artificial-Neural-Network/Single-Layer-Perceptron/Data Testing.xlsx")
df.head()
```

VISUALISASI DATA

```python
plt.figure(figsize=(10, 6))
sns.displot(data=df, x='PassengerId', kind='kde', color='black')
plt.title('Histogram Data')
plt.xlabel('PassengerId')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Pclass',  color='grey')
plt.title('Histogram Data')
plt.xlabel('Pclass')
plt.show()
```
