```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_excel("D:/visualisasi data/Artificial-Neural-Network/Single-Layer-Perceptron/Data Training.xlsx")
test = pd.read_excel("D:/visualisasi data/Artificial-Neural-Network/Single-Layer-Perceptron/Data Testing.xlsx")
```

```python
count_training = len(train)
count_testing = len(test)
```

```python
label =['Data Training', 'Data Testing']
size = [count_training, count_testing]
colors = ['blue', 'orange']
```

```python
plt.pie(size, labels=label, colors=colors, autopct='%1.1f%%', startangle=90)

# Menambahkan judul
plt.title('Perbandingan Data Training dan Data Testing')

# Menampilkan pie chart
plt.show()
```
