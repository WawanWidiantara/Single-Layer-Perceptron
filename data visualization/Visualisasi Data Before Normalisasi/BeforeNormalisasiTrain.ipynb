```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('D:/Matkul/JARINGAN SYARAF TIRUAN/TUGAS KELOMPOK/Data/Data Training.xlsx')
df
```

```python
survived_labels = ['Not Survived', 'Survived']

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='PassengerId', hue='Survived')
plt.title('Histogram Data')
plt.xlabel('PassengerId')
plt.ylabel('Jumlah')
plt.legend(title='Survived', labels=survived_labels)
plt.show()
```
